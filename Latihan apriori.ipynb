```python
import numpy as np
import  matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori 
```

```python
data = pd.read_excel('E:datamining-master/data.xls')
data.head()
```

```python
num_record= len(data)
print(num_record)
```

```python
records = []
for i in range(0,num_record):
    records.append([str(data.values[i,j]) for j in range(0,6)])
```

```python
association_rules = apriori(records, min_support=0.10, min_confidence=0.7, min_lift=1.2, min_length=1)
association_results = list(association_rules)
```

```python
print(len(association_results))
```

```python
print(association_results[0])
```

```python
results =[]
for item in association_results:
    pair = item[0]
    items = [x for x in pair]
    
    value0 = str(items[0])
    value1 = str(items[1])
    value5 = str(items[2])
    value2 = str(item[1])[:7]
    value3 = str(item[2][0][2])[:7]
    value4 = str(item[2][0][3])[:7]
    
    rows = (value0,value1,
            value5,
            value2,value3,value4)
    
    results.append(rows)
    
    label = ['title1','title2','title3',
             'support','confidence','lift']
    
    
    store_suggestion = pd.DataFrame.from_records(results,columns=label)
    
    print(store_suggestion)
```
